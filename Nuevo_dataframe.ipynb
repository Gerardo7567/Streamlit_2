```python
import pandas as pd
```

```python
df_hoteles = ('df_complete (1).csv')
df_hoteles = pd.read_csv(df_hoteles)
df_eventos = ('eventos_comerciales_bd.csv')
df_eventos = pd.read_csv(df_eventos)
df_eventos.head()
```

```python
df_hoteles['state'].value_counts()
```

```python
df_eventos = df_eventos.dropna()
df_hoteles["date"] =  pd.to_datetime(df_hoteles["date"], format="%m/%d/%Y")
df_eventos.rename(columns={'Date':'date'},
               inplace=True)
df_eventos["date"] =  pd.to_datetime(df_eventos["date"], format="%m/%d/%Y")
df_eventos = df_eventos.sort_values(by="date")

df_monterrey = df_hoteles[df_hoteles.state == 'Nuevo Leon']
```

```python
#df_complete = df_hoteles.join(df_eventos)
#df_complete = df_monterrey.merge(df_eventos)
#df_complete
df_complete = df_monterrey.merge(df_eventos,how='left', left_on='date', right_on='date')
df_complete.to_csv('Monterrey_dataset.csv')
df_complete
```
